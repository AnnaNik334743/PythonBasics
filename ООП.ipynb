# Что такое ООП

Python - объектно ориентированный язык, соответсвенно, он поддерживает ООП - объектно ориентированное программирование. 

ООП - это просто "новая ступень эволюции кода". Всё, что можно написать с помощью ООП, можно написать и без него с использованием функций, вложенных списков/словарей и их хитрого сплетения. Но зачастую через ООП писать проще, компактнее и быстрее.

Объект - это контейнер (коробка, в которую мы что-то складываем, и теперь эта коробка вместе с содержимым воспринимается нами как единое целое), состоящий из:

1) данных и состояния, то есть атрибутов (шерсть у моей кошки - белая, а глаза - голубые);

2) поведения, то есть методов (кошка мяукает, а собака лает).

Объекты определяются по принадлежности к классу. В Python всё является объектами!

```python
isinstance(4, float)  # функция, проверяющая, принадлежит ли число 4 классу float
```

```python
isinstance(4.0, float)
```

Класс - это шаблон для создания объектов. Если вернуться к аналогии с коробкой, то класс определяет её парметры: сколько она вмещает? что в неё можно класть? в каком порядке класть? есть ли внутри коробки секции и какие? может быть, в коробке уже что-то лежит вне зависимости от того, что мы еще хотим туда положить? 

Другое название объекта - экземпляр класса. На базе каждого класса можно создать сколько угодно экземпляров.

```python
import numpy as np

arr = np.array([1, 2, 3])  # экземпляр класса, где класс - нампаевские массивы, а экзмепляр - конкретный массив
isinstance(arr, np.ndarray)  
```

```python
arr.size  # атрибут 
```

```python
arr.reshape(-1, 1)  # метод
```

```python
len(arr)  # внешняя функция, не свзяаная с классом
```

Существуют так называемые парадигмы ООП:

1. Инкапсуляция - инициализация объекта (мы создали коробку!);
2. Наследование - дочерний класс радостно дублирует к себе всё, что имеет родительский класс;
3. Абстракция - отбрасывание всего лишнего и ненужного для работы. Нам неважно, как именно устроены классы под капотом;
4. Полиморфизм - переназначение методов. Помните, что дочерний класс забирает себе всё родительское? Так вот, это не приговор, его можно поменять в дальнейшем.

Еще нужно сказать про интерфейс. Это - совокупность методов и правил взаимодействия элементов системы. Другими словами, интерфейс определяет, как элементы будут взаимодействовать между собой. Благодаря интерфейсу, мы знаем, что можно делать с различными объектами.

```python
a = [1, 2, 3]  # список - объект
a.append(4)  # метод этого объекта
a.add(5)  # такого метода у данного класса нет, поэтому ошибка
```

# Классы

Поговорим про базовый синтаксис класса.

```python
class Cat:  # название класса принято писать с большой буквы

  num_of_legs = 4  # атрибуты класса
  num_of_eyes = 2  # они будут одинаковыми для всех экземпляров класса

  def __init__(self, name: str, color: str, fluffy: bool, lovely: bool=True):  # "магический метод", создающий экземпляр класса
    self.name = name  # атрибуты у каждого объекта будут свои, именно их мы передаем в скобочках
    self.color = color  # self указывает на то, что экземпляр работает сам с собой
    self.is_fluffy = fluffy  # мы в скобках передали какие-то значения, и через self объект их присваивает
    self.is_lovely = lovely  

  def get_color(self):  # метод, возвращающий свой атрибут
    return self.color

  def ask_food(self, time: int):  # метод, принимающий в себя сторонний параметр и работающий со своим атрибутом
    print(f"Meow! It's {time} o'clock, {self.name} is hungry!")

  @staticmethod  
  def _hiss():  # статический метод без self (ему ни методы, ни атрибуты класса не интересны)
    print("Hisss!")  # а еще он приватный 

  def purr(self):  # метод, вызывающий другой метод класса
    if np.random.randint(0, 2):
      print("Purrrrr...")
    else:
      self._hiss()
```

```python
def function(a, b, c=5):
  pass
```

```python
function(10, 4)
function(10, 4, c=7)
```

Нужно отметить, что у классов общая область видимости. Неважно, в каком порядке создавать функции внутри класса. Любой метод класса имеет доступ ко всем методам.

Про приватные методы. Это такие методы, которые вызываются только внутри класса (то есть их могут вызывать только другие методы этого же класса). Они начинаются с нижнего подчеркивания, и по факту являются программистким соглашением, потому что на самом деле в Python нет приватных методов. Просто программисты договорились, что методы, начинающиеся с нижнего подчеркивания, они не вызывают вне класса.

Антонимом приватных методов являются публичные. Их можно вызывать как внутри класса, так и вне.

Давайте создадим пару котиков и посмотрим, как они ведут себя.

```python
sam = Cat(name="Sam", color="black", fluffy=True)  # создаем экземпляры класса с определенными параметрами
lola = Cat(name="Lola", color="white", fluffy=True)  # эти параметры передаются в init
```

```python
print(sam.num_of_legs, lola.num_of_legs)  # атрибуты класса у всех экземпляров одинаковые
```

```python
print(sam.name, lola.name)  # атрибуты объектов будут разными
```

```python
sam.ask_food(8)  # метод класса
```

```python
print(sam.get_color(), lola.get_color())
```

```python
lola.purr()
```

```python
sam.purr()
```

# Наследование

Наследование происходит по следующему принципу: пусть у нас есть родительский класс со своими атрибутами и методами. Создадим на его базе дочерний. Дочерний класс будет иметь такие же атрибуты и методы, как родительский. Но особого смысла в двух идентичных классах с разными именами нет, поэтому параметры дочернего класса можно (и нужно) переназначить. Посмотрим на примере.

```python
class Rectangle:
  def __init__(self, a, b):  # прямоугольник однозначно задается длинами двух его сторон
    self.width = min(a, b)
    self.length = max(a, b)
    self.p = self._get_p()
    self.s = self._get_s()

  @staticmethod
  def definition():
    print("This is a rectangle")

  def _get_p(self):  # периметр
    return 2 * (self.width + self.length)

  def _get_s(self):  # площадь
    return self.width * self.length
```

```python
rect = Rectangle(5, 10)
rect.definition()
```

```python
print(rect.p, rect.s)
```

```python
class Square(Rectangle):
  def __init__(self, a):  # переназначим init, потому что для квадрата нет смысла передавать две стороны
    self.width = a
    self.length = a
    self.p = self._get_p()
    self.s = self._get_s()

  def new_definition(self):
    print("This is a square")
```

```python
square = Square(5)
square.definition()  # этот метод был у родительского класса
```

```python
print(square.p, square.s)  # и эти методы тоже от родителя
```

```python
square.new_definition()  # а этого метода у родительского класса не было
```

```python
rect.new_definition()  # изменения в дочернем классе не влияют на родительский
```
