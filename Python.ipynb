Пример кода на Python - сортировка пузырьком

```python
import typing as tp
from copy import deepcopy


def bubble_sort(some_list: tp.List, reverse=False) -> tp.List:
    """

    :param some_list: initial list for sorting
    :param reverse: if True, the list will be sorted backwards
    :return: sorted list

    """
    my_list = deepcopy(some_list)
    for i in range(len(my_list) - 1):
        flag = False
        for j in range(len(my_list) - 1):
            if not reverse:
                if my_list[j] > my_list[j+1]:
                    flag = True
                    my_list[j], my_list[j+1] = my_list[j+1], my_list[j]
            elif my_list[j] < my_list[j+1]:
                flag = True
                my_list[j], my_list[j+1] = my_list[j+1], my_list[j]
        if not flag:
            break
    return my_list


if __name__ == '__main__':
    array = [1, 9, -55, 3, 6, 18, 0, 0, -2]
    print(f"Initial array: {array}")
    print(f"Sorted array: {bubble_sort(array)}")
    print(f"Array sorted backwards: {bubble_sort(array, reverse=True)}")
```

Комментарий пишется после решетки:

```python
def get_digits(n: int) -> tp.List[int]: # функция для нахождения разрядов числа
  list_of_dig = [] # создаем пустой список
  while n != 0: # пока разряды остались
    list_of_dig.append(n % 10) # берем разряд
    n //= 10 # отрезаем уже взятый разряд
  return list_of_dig[::-1] # разворачиваем список разрядов


print(get_digits(12131))
```

# Циклы

Цикл нужен для многократного повторения одних и тех же операций, чтобы мы не писали 100500 раз одинаковые строки кода. В большинстве языков программирования есть три типа циклов: с предусловием (while), со счетчиком (for) и с постусловием (until). В Python есть только первые два.


Начнем с цикла с предусловием. Пока а, делай б. Напишем программу, которая будет считать сумму всех введеннных пользователем чисел, если они делятся на два без остатка. Когда пользователь введет 0, программа завершится.

```python
summa = 0
n = int(input())
while n != 0:
  if n % 2 == 0: 
    summa += n
  n = int(input())
print(summa)
```

Чуть-чуть поменяем условие. Теперь вводить можно только числа, большие нуля. Но обратите внимание на блок else после тела цикла. Когда мы выйдем из цикла, то зайдем в него.

```python
summa = 0
n = int(input())
while n > 0:
  if n % 2 == 0: 
    summa += n
  n = int(input())
else: 
  print("Ага, импостер!!")
print(summa)
```

В случае цикла с предусловием есть такая вещь, как бесконечный цикл. Это очень плохо и обычно случается, когда вы накосячили с условием. Вот мы и подошли к циклу for. В некоторых языках программирования он строится по следующему принципу: берем какую-нибудь переменную-счетчик, устанавливаем для неё начальное значение, конечное и шаг. Похоже на функцию range(), не правда ли? 

Но цикл в Питоне работает несколько иначе, по принципу 

for <переменная> in <итерируемый объект>: 

> <тело цикла>

Что такое итерируемый объект? Это любая последовательность. Например, это может быть набор переменных, список, состоящая из символов строка, заданная функцией range() последовательность и т.п. Переменная-счетчик на самом деле не является счетчиком! В ходе выполнения цикла меняется не она, а автоматически создающийся "указатель" в итерируемом объекте. И переменная всего лишь принимает то значение, на котором на данной итерации цикла оказался указатель. Переменную можно менять в теле цикла, в Питоне никаких санкций из-за этого не будет.

Теперь, когда вы окончательно запутались, где счетчик, а где указатель, можете просто забыть то, что тут написано, и посмотреть примеры.

```python
for i in range(4):
  print(i)
```

```python
for j in range(5):
  print(j if j % 3 == 0 else None)
  j = 100 # C++ или Паскаль стали бы ругаться на это. Питону все равно, потому что j - это не счетчик
```

```python
for letter in "JhgHghsatgGCgfcofeioi":
  if letter in "eioa": # проверяем, содержится ли текущий символ в указанной последовательности
    print(letter)
```

Еще несколько полезных методов и функций для работы со списками. Пусть наш список имеет имя a.

* len(a) - длина списка
* a.append(x) - добавляет элемент x в конец списка
* a.extend(seq) - добавляет в конец списка все элементы последовательности seq
* del a[i] - удаляет элемент с индексом i. Если такого элемента не окажется, выдаст ошибку

Срезы - это когда мы берем не весь список, а только интересующую нас часть списка. some_list[begin, end, step] - возьмет значения списка из диапазона от begin до end-1 с шагом step. Шаг по умолчанию равен одному.

Поговорим о разных полезных методах для работы со строками. Пусть наша строка имеет имя s.

```python
from string import printable as pr

s = input()
if len(s) < 16:
    print('No')
else:
  flag = True
  symbols = [False, False, False, False]
  for i in s:
      if i in pr[:10]:
          symbols[0] = True
      elif i in pr[10:36]:
          symbols[1] = True
      elif i in pr[36:62]:
          symbols[2] = True
      elif i in pr[62:]:
          symbols[3] = True
  for j in symbols:
      if not j:
          print('No')
          flag = False
          break
  if flag:
    print('Yes')
```

Потому что тело главной программы не видит, что происходит в теле функции. Область видимости переменных - важное понятие, которое необходимо усвоить. Программа видит только то, что относится к ней самой. По этой же причине вы не можете обратиться извне к переменной, объявленной внутри цикла. 

В Питоне по умолчанию все переменнные локальные. В некоторых языках, например в Bash, все по умолчанию глобальные. Локальные - значит, обратиться к ним извне не получится. Глобальные - значит, их видно вообще ото всюду.

Правило простое: если вы хотите поменять переменную внутри функции - передайте её в качестве аргумента.

Если вам для каких-то целей все-таки потребуется обратиться к переменной главной программы из функции (обычно для глобального счетчика) - пропишите перед ней слово global. Это позволит изменять переменную на всех уровнях.

```python
def делайте_так():
  global внешняя_переменная 
  внешняя_переменная += 10
  print(внешняя_переменная)


внешняя_переменная = 100
делайте_так()
```

Помимо global есть еще nonlocal. Оно используется, когда у нас функция-в-функции и мы хотим обратиться из внутреннней функции ко внешней.

```python
def bar():
  var = 11
  def foo():
    nonlocal var
    var = 100
  foo()
  return var


var = 10
print(bar())
print(var)
```

Рекурсия - это когда функция вызывает сама себя.

ojI6AxR.jpg

```python
def rec(i):
  if i >= 100:
    return i
  return rec(i + 5)


rec(3)
```

В качестве примера напишем программу рекурсивной проверки на палиндром.

```python
def palindrom(s):
  if len(s) <= 1:
    return True
  if s[0] != s[-1]:
    return False
  return palindrom(s[1:-1])
```

```python
print(palindrom("арозаупаланалапуазора"))
print(palindrom("urrf"))
```

Если мы пишем рекурсию, то всегда должно достигаться какое-либо состояние, когда из рекурсии можно выйти. То есть, рано или поздно функция должна достигнуть строчки return.

Проблема рекурсии в том, что если вы не прописали условие выхода из нее, получится что-то вроде бесконечного цикла. Подумайте, что не так со следующим кодом.

```python
def rec(i):
  if i >= 100:
    return i
  return rec(i + i % 2)


rec(3)
```

Не всегда имеет смысл использовать рекурсию, иногда проще обойтись циклом. Приведу пример. Пусть у нас есть матрешка. Мы не знаем, сколько в ней еще матрешек, но точно знаем, что последняя из них пустая. Есть два сценария:

1. Нам нужно найти пустую матрешку
2. Нам нужно найти пустую матрешку, положить внутрь записку и закрыть все матрешки

Первую задачу логичнее решать с помощью цикла while. Подумайте, как. Во втором же случае удобнее решать через рекурсию, потому что она позволит "откатиться" назад.

Псевдокод:

```python
def матрешки(матрешка):
  открыть матрешку
  if матрешка пустая:
    положить записку
    закрыть матрешку
    return закрытая_матрешка
  матрешки(открытая матрешка)
  закрыть матрешку
  return закрытая_матрешка
```

Скажу честно: за полгода работы с Питоном рекурсия пригодилась мне один раз, и то пришлось переписывать на цикл, потому что в Питоне стоит ограничение глубины рекурсии (если функция вызвала саму себя слишком много раз, даже если вы так и рассчитывали, программа выдаст ошибку).

Модуль random (https://pythonworld.ru/moduli/modul-random.html) - незаменимая вещь. Некоторые полезные функции из него:

```python
import random


x = random.randint(5, 10) # вернет рандомное целое число из указанного в скобках диапазона
print(x)
y = random.random() # вернет случайное вещественное значение от 0 до 1
print(y)
l = random.choice("123ghj456as") # вернет рандомный символ из последовательности
print(l)
lst = [21, 67, 32, 1, 0, 1]
random.shuffle(lst) # перемешает последовательность
print(lst)
```

Доктесты - это такие небольшие тесты, которые удобно использовать при отладке программы. Посмотрите на то, как они пишутся:

```python
def R_Number(s):
    """
    Tests to see if number is int or float.
    >>> R_Number("122") # ввод
    True # предполагаемый вывод
    >>> R_Number("102123124840939824.1")
    True
    >>> R_Number("-9298.212")
    True
    >>> R_Number("012.1")
    False
    >>> R_Number("-0002")
    False
    >>> R_Number("13.00021")
    True
    >>> R_Number("-0.9")
    True
    >>> R_Number("-0..9")
    False
    >>> R_Number("-0.09")
    True
    >>> R_Number("-0,09")
    False
    """
    if s[0] == "-":
        s = s[1:]
    k_points = 0
    for i in s:
        if not i.isdigit():
            if i == "." and k_points == 0:
                k_points += 1
            else:
                return False
    if len(s) > len(str(float(s))):
        return False
    return True
```
